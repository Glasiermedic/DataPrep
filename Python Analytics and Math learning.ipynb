```python
# @title
#Order of Operations - PEDMAS

# prompt: function that simplifies the expression
ord_op_list = []
#import sympy

def simplify_expression(expression):
  """
  Simplifies the given expression using SymPy.

  Args:
    expression: A string representing the expression to simplify.

  Returns:
    The simplified expression as a string.
  """

  # Convert the expression to a SymPy expression object.
  sympy_expression = sp.sympify(expression)

  # Simplify the expression.
  simplified_expression = sp.simplify(sympy_expression)

  # Convert the simplified expression back to a string.
  return str(simplified_expression)

# Test the function.
for expression in Order_operations_examples:
  #print("Original expression:", expression)
  #print("Simplified expression:", simplify_expression(expression))
  #print()
  ord_op_list.append(simplify_expression(expression))

#checking for issues with lists - bug check
#print(len(ord_op_list))
#print(len(Order_operations_examples))

print("PEDMAS is the order of operations for math")
print("Parentheses () first")
print("Exponents ^ are addressed second")
print("Then do multiplcation * and/or division / start from left to right")
print("Then do addition + and subtraction - start from left to right")
ord_op = pd.DataFrame({'Id': Id_col, 'examples': Order_operations_examples, 'solution': ord_op_list})
print(ord_op)
```

## Derivatives

```python
# @title
derr_def = """ Derivatives tell you the slope of a function at any given point,
the rate of change. """

print(derr_def)

print("an example is: ")

def derivative_x(f, x, step_size):
  m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
  return m

def my_function(x):
  return x**2 + 2*x + 1

slope_at_x = derivative_x(my_function, 2, 0.0001)
print("slope at x = 2: ", slope_at_x, "for x**2 + 2*x + 1")

```

## Limits and Calculating Derivatives

```python
# @title
lim_def = """ A limit is a term used to help in specific functions.
Whenever an input is forever increased or decreased the output approaches a
number but never actually reaches it we what is called a limit."""

print(lim_def)

print("example of function that generates 2 limits 10 and -10: y = (1/x) + 10")

def f(x):
  return (1/x)+10

def g(x):
  return (x/x)*10

# Generate numerical data for plotting
x_vals = np.linspace(-5, 5, 100)  # Create an array of x-values
y_vals = f(x_vals)  # Calculate corresponding y-values
y2_vals = g(x_vals) # Calculate corresponding y-values
fig = plt.figure()
plt.plot(x_vals, y_vals, color = 'blue') # Use numerical data for plotting
plt.plot(x_vals, y2_vals, color = 'red') # Use numerical data for plotting
plt.grid()
plt.show()


```

```python
# @title
limder_def = """ We can use the principles around a limit to calculate the
derivative of a function as well. """

print(limder_def)

# "x" and the step size "s" for calculation

x,s = symbols('x s') #establish symbolic variables

f = x**2  # define the function

# slope is calculated at rise over run and we can add our gap "s" into formula
# y_1 - y_2 / x_1 - x_2

slope_f = (f.subs(x, x + s) - f) / ((x+s) -x)

result = limit(slope_f, s, 0)
print(result)



```

## Chain Rule

```python
# @title
#  y = x**2 + 1
#  z = y**3 - 2

# z = (x**2 + 1)**3 -2

print(" To explain chain rule we can use two formulas: ")
print("y = x**2 + 1")
print("z = y**3 - 2")

print("as they are linked we can make it one : z = (x**2 + 1)**3 -2")

print("we can get the derivative of z with respect to x by the following:")
print()
print("""z = (x**2 + 1)**3 -2
dz_dx = diff(z,x)""")

z = (x**2 + 1)**3 -2

dz_dx = diff(z,x)
print()
print("derivative:")
print(dz_dx) # 6*x*(x**2 + 1)**2
print()

print (""" we can also get the derrivative by multiplying the dz_dy and dy_dx
derivatives to get the complete derivative""")
# derivative for first function
_y = x**2 +1 #can't reuse y due to variable clash
dy_dx = diff(_y)

# derivative for second function
z =  y**3 -2
dz_dy = diff(z)

dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y,_y))
print()
print("""Comparing the results of both methods:
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
""", dz_dx_chain)
print("""dz_dx_no_chain = diff(z.subs(y,_y))
""",
      dz_dx_no_chain)
#Now calculate derivative with and without
#chain rule, substitute y function

print()
print("""We can use chain rule to help build neural networks and train them
so that they have proper weights and biases""")
```

```python
# @title
Ingrl_def = """ Integrals are the area under a curve or line in a given range of
 a given axis.

 Rectangle approximation of integral:
  We can approximate what the area of curve is by using rectangles
 of equal width and a height where the mid point of the top edge touches the
 line or curve, as we increase the number of rectangles and reducing the width
 of each rectangle we can get a more accurate approximation of the area.
 This method is called the Trapezoidal Rule, which works well for simple curves
 and less precise needs but the most powerful and precise method is to use
 integrals.
 """

print(Ingrl_def)

def approximate_integral(a, b, n, f):
  delta_x = (b-a)/n
  total_sum = 0

  for i in range(1, n+1):
    midpoint = 0.5 * (2 * a + delta_x * (2*i - 1))
    total_sum += f(midpoint)

  return total_sum * delta_x

def my_function(x):
  return x**2 + 1

input_var = input("Enter number of rectangles: ")
if int(input_var) < 1000000:
  area = approximate_integral(a = 0, b = 1, n = int(input_var), f = my_function)
else:
  print("okay you did good but the function freaks out over 1 million so ...")
  input_var = 1000000
  area = approximate_integral(a = 0, b = 1, n = 1000000, f = my_function)
print("area: ", area, " with ", input_var, """rectangles.  As we pick more and more
rectangles we approach 1.33 or 4/3""")
```

# Chapter 2: Probability

Probability is how strongly we believe an event will
happen, often expressed as a percentage.

Example Probability problems:

How likely will I get 6 heads in 10 fair coin flips?

What are the odd of losing a city counsel election?

Will my flight be delayed?

How certain can you be that a product is defective?

Probabilities are best expressed as percentages or ratios.   Programming languages default to decimals to represent the probability where 100% is 1.0.   Because of this for all events when it not possible to have multiple outcomes at the same time the sum of all probable outcomes must equal 1 or 100%.



Likelihoods are about how frequently of events that already occured and does not follow the limits of probability (the sum of all probabilities when one one outcome is possible must equal 100%)

The functions for probability are:

P(X) = n.
n being decimal of the percentage
P(not X) = 1-n

P(X) = O(X)/(1 + O(X))

where O(X) is the odds: 7:3, 7/3

O(X) = P(X)/(1-P(X)) the probability of X happening divided by X not happening (sum of all other options).

7/3 is not a violation of our probability as it can also be written as 7 of 10 over 3 of 10

**Joint Probability** is defined as two or more events that are independent in action but included in the probability of a specific outcome for each event. For example the probability of getting a heads on a coin and also the a single die rolling a 4.

## Union Probabilities

**Union Probabilities** are defined as the probability of A OR B occuring

When A cannot happen at the same time as B then this is mutually exclusive and be written and solved as such:

P(A OR B) = P(A) + P(B)
A = 1/6 B = 1/6  = 2/6 = 1/3

When A can happen at the same time as B this is nonmutually exclusive.  IF A is Heads on a coin and B is 6 on the dice:

*H1* *H2* *H3* *H4* *H5* *H6* T1 T2 T3 T4 T5 *T6*

A = 1/2= 3/6
B = 1/6  

but  A OR B != 4/6

A OR B = 7/6

The rule for non exclusive A OR B Union probablity is *Sum Rule of Probability*:  to sum the probabilities and then subtract the joint probability from it.  

P(A OR B) = P(A) + P(B) -( P(A) x P(B))

1/2 + 1/6 - (1/2 x 1/6) = 4/6 -(1/12) = 7/12  

This will work for mutually exclusive and non mutually exclusive as well.

## Conditional Probablilities

Conditional probabilities are defined as one event occuring given that another fact or condition is true/occured, not neccasarily at the same time.  Something to be careful about is false logic.   5% of americans have cancer and 51% of the US population is female.  

If we try and calculate the probability of having cancer if you are male we might be tempted to just multiply .05 x .49 (24.5%) that is a misleading statistic, we actually are saying that despite the population being close to 50/50 Male/Female the percentage of cancer patients that are male is 24.5% not that if you are male the probability that you also have cancer.  And to make it worse that is not even accurate for the actual probability of finding out you are male (49% of population) given you already have cancer (5% of population)

Another example unsubstantiated news reports like 86% of American Plane crash victims own a boat. ### Bayes Theorem

Allows us to flip the conditional probability from B given A to A given B.  

Bayes Theorem : the relation between a conditional probability and its reverse form can be used to if we have an observed event and the probability of that event occuring.

p(**A&B**|B) = p(B&A|A) x p(A) / p(B)

p(B&A|A) = p(**A&B**|B) x p(B) / p(A)

this way of writing the theorem is often simplified to

p(**A|B**) = p(B|A) x p(A) / p(B)    
*probability of A happening given B is equal to the probability of B happening given A mulitpled by probability of A divided by probability of B*

p(B|A) = p(**A|B**) x p(B) / p(A)
*probability of B happening given A is equal to the probability of A happening given B mulitpled by probability of B divided by probability of A*

... So if I know P(A|B) and p(B) I can get the p (B|A) and vise versa


A simple example is as follows: There is a 40% chance of rain on Sunday. If it rains on Sunday, there is a 10% chance of rain on Monday. If it did not rain on Sunday, then there is an 80% chance that it will rain on Monday.

"Raining on Sunday" is event A, and "Raining on Monday" is event B.

P( A ) = 0.40 = Probability of **rain on Sunday**.

P( A` ) = 0.60 = Probability of no rain on Sunday.

**P( B | A ) = 0.10** = Probability of **rain on Monday, if it rained on Sunday**.

P( B` | A ) = 0.90 = Probability of no rain on Monday, if it rained on Sunday.

P( B | A` ) = 0.80 = Probability of **rain on Monday, if it did not rain on Sunday.**

P( B` |A` ) = 0.20 = Probability of no rain on Monday, if it did not rain on Sunday.

We need to know the probability that it rained on Sunday given that it rained on Monday.  (I know that is silly but what if you were gone Sunday and want to know if your plants got enough water over the last 48 hrs)

We need to get one more probability P(B) which is P(A) x P(B|A) + P(A`) x P(B|A`)
P(B) = 0.40 x .10 + .60 x .80 = .52

Then we can put those into our equation:

P(A|B) = P(B|A) x P(A)/P(B)  

**P(A|B) = (.10 x .40) / .52 = .769  or 76.9% chance it rained on Sunday if it rained on Monday.**

## Binomial Distribution
is one of the probability distributions commonly used.  Binomial distribution measures the likelihood of k successes occuring out of n trials given probability of p.   

This comes up a lot when we recognize that we are limited by time, money, people, supplies to perform a large number of tests and need to know the likelihood you will get the about the same results if you did the same, less or more tests.  And to be clear binomial distributions produce a table or a chart that shows you the probability of getting the outcome you predict with n number of trial (repeat tests) given the probability of the result occuring absolutely.
